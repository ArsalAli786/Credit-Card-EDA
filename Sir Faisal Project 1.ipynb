```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.isnull().sum()
```

### ● How many transactions are fraudulent, and how many are legitimate? 

```python
df['Class'].value_counts()
```

### ● What percentage of transactions are fraudulent? 

```python
f'{(df['Class'].sum() / len(df) * 100) :.2f}%'
```

### ● What are the minimum, maximum, mean, and median values for numerical columns like Amount?

```python
print(df['Amount'].mean())
print(df['Amount'].median())
print(df['Amount'].max())
print(df['Amount'].min())
```

### ● What is the maximum transaction amount in the dataset, and is it fraudulent?

```python
max = df['Amount'].max()
max_transaction = df[df['Amount'] == max]
fraudulent = max_transaction['Class'].values[0] == 1
print(max)
if fraudulent == False:
    print('No')
```

### ● Can we create a bar chart showing the count of fraudulent vs. legitimate transactions? 

```python
sns.countplot(x = 'Class', hue = 'Class', data = df)
plt.xticks([0, 1], ['Legitimate', 'Fraudulent'])
plt.xlabel('Transactions')
plt.ylabel('Number of Transactions')
plt.title('Legitimate vs Fraudulent Transactions')
plt.show()
```

### ● What does the histogram of transaction amounts look like?

```python
plt.hist(df['Amount'], bins=50, color='blue', edgecolor='black')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Transaction Amounts')
plt.show()
```

### ● Can we use a heatmap to visualize the correlation between numerical features?

```python
correlation_matrix = df.corr()
plt.figure(figsize = (14, 10))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', fmt = '.1f', linewidths = 0.1)
plt.title('Correlation Heatmap between Numerical Features')
plt.tight_layout()
plt.show()
```
